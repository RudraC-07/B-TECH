### 01) Write a function to calculate BMI given mass and height. (BMI = mass/h**2)

```python
def Calculate_BMI(m,h):
    bmi = m/(h**2)
    print("Your BMI : ",bmi)

m = int(input("Enter mass : "))
h = int(input("Enter height : "))
Calculate_BMI(m,h)
```

### 02) Write a function that add first n numbers.

```python
def Add_n_Numbers(n):
    s = 0
    for i in range(1,n+1):
        s = s+i
    print("Sum of first n numbers : ",s)

n = int(input("Enter a number : "))
Add_n_Numbers(n)
```

### 03) Write a function that returns 1 if the given number is Prime or 0 otherwise.

```python
def Prime_or_Not(n):
    c = 0
    for i in range(1,n+1):
        if(n % i == 0):
            c += 1
    if(c == 2):
        return 1
    else:
        return 0

n = int(input("Enter a number : "))
Prime_or_Not(n)
```

### 04) Write a function that returns the list of Prime numbers between given two numbers.

```python
def List_Prime_Numbers(a,b):
    lst = []
    for i in range(a,b+1):
        if Prime_or_Not(i):
            lst.append(i)
    return lst

a = int(input("Enter 1st number : "))
b = int(input("Enter 2nd number : "))
List_Prime_Numbers(a,b)
```

### 06) Write a function that returns the sum of all the elements of the list.

```python
def List_Add():
    s = 0
    l1 = input("Enter elements : ").split(" ")
    l2 = [int(i) for i in l1]
    for i in l2:
        s += i
    print("Sum of elements : ",s)

List_Add()
```

### 07) Write a function to calculate the sum of the first element of each tuples inside the list.

```python
def Add_First_Ele():
    s = 0
    l1 = [(1,2,3),(4,5,3),(8,5,2)]
    s = sum(i[0] for i in l1)
    print("Sum of first tuple elements : ",s)

Add_First_Ele()
```

### 12) Write a function to check whether the given string is Pangram or not.
#### hint: Pangram is a string containing all the characters a-z atlest once.
#### "the quick brown fox jumps over the lazy dog" is a Pangram string.

```python
from string import ascii_lowercase

def Pangram_or_Not():
    str1 = "the quick brown fox jumps over the laz dog"
    str2 = ascii_lowercase
    for i in str2:
        if i not in str1:
            return False
    return True

print(Pangram_or_Not())
```
